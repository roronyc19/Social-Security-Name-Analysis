```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### Create dataframe from 1880-2020

yearInt=1880
locationStringFront = "/Users/tom_sherman/Desktop/Names SS/names/yob"
locationStringBack = ".txt"
SSDataBaseList = []
while yearInt<2021 :
    yearString = str(yearInt)
    locationStringFront+=yearString
    locationStringFront+=locationStringBack
    df = pd.read_csv(locationStringFront)
    df.columns = ['Name', 'Sex', 'Number of People']
    df["Year"]=""
    df.loc[:, "Year"]=yearInt
    df["Proportion of Population"] = df["Number of People"].div(df["Number of People"].sum())
    SSDataBaseList.insert(yearInt-1880, df)
    yearInt+=1
    locationStringFront= "/Users/tom_sherman/Desktop/Names SS/names/yob"
    dataFrameName="Data"
result = pd.concat(SSDataBaseList)
resultM = result[result["Sex"]=="M"]
resultF = result[result["Sex"]=="F"]
```

```python
#This code allows you to select a name and sex and see the resulting name graph


name = input("Input Desired Name: ")
print("Name selected: " + name)
sex = input("Select Desired Sex. Type 'Male' or 'Female': ")
print("Selected sex is " + sex)
if sex == "Male":
    df = resultM
if sex == "Female":
    df = resultF
dfNew = df[df["Name"]==name]
plt.plot(dfNew["Year"], dfNew["Number of People"])
plt.xlabel("Year")
plt.ylabel("Number of People")
plt.title("Number of People with the Name " + name + " From 1880-2020")
```

```python
#This code allows for male-female displays on the same graph

name = input("Input Desired Name: ")
print("Name selected: " + name)
#Male line
dfNewM = resultM[resultM["Name"]==name]
#Female line
dfNewF = resultF[resultF["Name"]==name]
plt.plot(dfNewM["Year"],dfNewM["Number of People"], label="Male")
plt.plot(dfNewF["Year"],dfNewF["Number of People"], label="Female")
plt.xlabel("Year")
plt.ylabel("Number of People")
plt.legend()
plt.title("Number of People of the Name " + name + " Male vs Female From 1880-2020")
```
